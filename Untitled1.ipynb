```python
import h5py
import numpy as np

# File path
file_path = r"C:\Users\risha\Desktop\pcm_project\pcm_2d_dataset.h5"

f = h5py.File(file_path, "r")

T  = f["T"][:]
lf = f["lf"][:]
ux = f["ux"][:]
uy = f["uy"][:]

t = f["t"][:]
x = f["x"][:]
y = f["y"][:]
params = f["params"][:]

f.close()

print("✅ Loaded:", T.shape)
print(f"Cases: {T.shape[0]}, Time steps: {T.shape[1]}, Grid: {T.shape[2]}x{T.shape[3]}")
```

```python
print("Preparing data (this may take a minute)...")

X_list = []
Y_list = []

for c in range(T.shape[0]):
    T_hot = params[c, 0]
    
    for ti in range(T.shape[1]):
        for j in range(T.shape[2]):
            for i in range(T.shape[3]):
                
                X_list.append([x[i], y[j], t[ti], T_hot])
                Y_list.append([T[c, ti, j, i], lf[c, ti, j, i], 
                              ux[c, ti, j, i], uy[c, ti, j, i]])

X = np.array(X_list, dtype=np.float32)
Y = np.array(Y_list, dtype=np.float32)

print(f"✅ X shape: {X.shape}")
print(f"✅ Y shape: {Y.shape}")
print(f"Total samples: {X.shape[0]:,}")
```

```python
# Save normalization parameters for later use
X_mean = X.mean(axis=0)
X_std  = X.std(axis=0) + 1e-6

Y_mean = Y.mean(axis=0)
Y_std  = Y.std(axis=0) + 1e-6

X = (X - X_mean) / X_std
Y = (Y - Y_mean) / Y_std

print("✅ Data normalized")
print(f"X range: [{X.min():.2f}, {X.max():.2f}]")
print(f"Y range: [{Y.min():.2f}, {Y.max():.2f}]")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, Y_train, Y_val = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

print(f"✅ Train samples: {X_train.shape[0]:,}")
print(f"✅ Val samples: {X_val.shape[0]:,}")
```

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
import numpy as np

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"✅ Device: {device}")

if device == "cuda":
    print(f"   GPU: {torch.cuda.get_device_name(0)}")
    torch.backends.cudnn.benchmark = True
    torch.set_float32_matmul_precision('high')  # Mixed precision

# Convert to tensors (Foolproof method for both numpy arrays and tensors)
X_train = torch.as_tensor(X_train, dtype=torch.float32).to(device)
Y_train = torch.as_tensor(Y_train, dtype=torch.float32).to(device)
X_val = torch.as_tensor(X_val, dtype=torch.float32).to(device)
Y_val = torch.as_tensor(Y_val, dtype=torch.float32).to(device)

print(f"✅ Data on {device}")

# WIDER + DEEPER Model
class SirenLayer(nn.Module):
    def __init__(self, in_f, out_f, w0=30.0, is_first=False):
        super().__init__()
        self.in_f = in_f
        self.w0 = w0
        self.linear = nn.Linear(in_f, out_f)
        self.is_first = is_first
        self.init_weights()
    
    def init_weights(self):
        with torch.no_grad():
            if self.is_first:
                self.linear.weight.uniform_(-1 / self.in_f, 1 / self.in_f)
            else:
                self.linear.weight.uniform_(-np.sqrt(6 / self.in_f) / self.w0, 
                                           np.sqrt(6 / self.in_f) / self.w0)
    
    def forward(self, x):
        return torch.sin(self.w0 * self.linear(x))

class PCM_SIREN(nn.Module):
    def __init__(self, hidden=512, n_layers=8):
        super().__init__()
        
        self.net = nn.ModuleList()
        self.net.append(SirenLayer(4, hidden, is_first=True))
        
        for _ in range(n_layers):
            self.net.append(SirenLayer(hidden, hidden))
        
        self.final = nn.Linear(hidden, 4)
        
        with torch.no_grad():
            self.final.weight.uniform_(-np.sqrt(6 / hidden) / 30.0, 
                                       np.sqrt(6 / hidden) / 30.0)
    
    def forward(self, x):
        for layer in self.net:
            x = layer(x)
        return self.final(x)

print("✅ SIREN model architecture ready")
```

```python
train_ds = TensorDataset(X_train, Y_train)
val_ds = TensorDataset(X_val, Y_val)

# Removed pin_memory=True because data is already on the GPU!
train_dl = DataLoader(train_ds, batch_size=16384, shuffle=True)
val_dl = DataLoader(val_ds, batch_size=16384)

print(f"✅ DataLoader ready (batch=16384)")

def data_loss(pred, target):
    return torch.mean((pred - target)**2)

def physics_loss_fast(model, X_batch):
    """Lightweight physics"""
    n = min(2048, X_batch.size(0))
    X = X_batch[:n].requires_grad_(True)
    
    pred = model(X)
    T = pred[:, 0:1]
    lf = pred[:, 1:2]
    
    # Gradient penalty (smooth solutions)
    grads = torch.autograd.grad(T.sum(), X, create_graph=True)[0]
    grad_penalty = torch.mean(grads**2)
    
    # LF bounds
    lf_penalty = torch.mean(torch.relu(-lf)**2 + torch.relu(lf - 1)**2)
    
    return grad_penalty + lf_penalty

print("✅ Loss functions ready")
```

```python
import matplotlib.pyplot as plt
import torch
from torch.amp import autocast, GradScaler

def train_single_model(model_id, n_epochs=500):
    print(f"\n{'='*70}")
    print(f"TRAINING MODEL {model_id}/3 (Plateau Breaker Edition v2)")
    print(f"{'='*70}\n")
    
    # Fresh model initialization
    model = PCM_SIREN(hidden=512, n_layers=8).to(device)
    print(f"Parameters: {sum(p.numel() for p in model.parameters()):,}")
    
    # Optimizer with safe base LR and low weight decay
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-5)
    
    # SMART SCHEDULER: Patience=3 means if val_loss is stuck for 3 checks (30 epochs), LR drops by half.
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='min', factor=0.5, patience=3, min_lr=1e-7
    )
    
    scaler = GradScaler('cuda') 
    
    best_val = float("inf")
    val_history = []
    
    for epoch in range(1, n_epochs + 1):
        
        # Smooth continuous curriculum for physics and temperature weights
        progress = epoch / n_epochs
        w_phys = 0.001 + (0.049 * progress)
        w_T = 15.0 + (10.0 * progress)
        
        model.train()
        
        for X_b, Y_b in train_dl:
            optimizer.zero_grad()
            
            with autocast('cuda'):
                pred = model(X_b)
                
                # BUMPED UP LIQUID FRACTION WEIGHT (5.0 -> 20.0)
                L_data = (
                    w_T * torch.mean((pred[:,0] - Y_b[:,0])**2) +
                    20.0 * torch.mean((pred[:,1] - Y_b[:,1])**2) +  
                    0.05 * torch.mean((pred[:,2] - Y_b[:,2])**2) +
                    0.05 * torch.mean((pred[:,3] - Y_b[:,3])**2)
                )
                
                L_phys = physics_loss_fast(model, X_b)
                loss = L_data + w_phys * L_phys
            
            scaler.scale(loss).backward()
            scaler.unscale_(optimizer)
            
            # Tighter gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), 0.5)
            
            scaler.step(optimizer)
            scaler.update()
            
        # Validation Check (Every 10 epochs)
        if epoch % 10 == 0:
            model.eval()
            val_loss = 0
            with torch.no_grad():
                for X_b, Y_b in val_dl:
                    pred = model(X_b)
                    val_loss += data_loss(pred, Y_b).item()
            
            val_loss /= len(val_dl)
            val_history.append((epoch, val_loss))
            
            # Step the smart scheduler based on Validation Loss
            scheduler.step(val_loss)
            
            lr = optimizer.param_groups[0]['lr']
            print(f"E{epoch:03d} | Val:{val_loss:.5f} | LR:{lr:.1e} | w_phys:{w_phys:.4f}", end="")
            
            if val_loss < best_val:
                best_val = val_loss
                torch.save(model.state_dict(), f"model_{model_id}.pt")
                print(" ✅")
            else:
                print()
    
    print(f"\n✅ Model {model_id} done | Best: {best_val:.5f}\n")
    return best_val, val_history

# Train all 3 ensemble models
results = []
for i in range(1, 4):
    best, hist = train_single_model(i, n_epochs=500)
    results.append((best, hist))

print("\n" + "="*70)
print("INDIVIDUAL MODEL RESULTS:")
for i, (best, _) in enumerate(results, 1):
    print(f"  Model {i}: {best:.5f}")
print("="*70)
```
